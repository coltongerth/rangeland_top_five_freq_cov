```python
import pandas as pd
from google.colab import drive
drive.mount('/content/gdrive')
```

Inputs

```python
number_of_top = 5
frequency_column = 'SumOfFrequency'
abs_cov_column = 'AvgOfAvgOfLFAbsCov'
# The name of the file with the extension that is put into the local directory
data_path = "NC_Species_Summary_query.xlsx"
# Just the filename with the .xlsx removed, feel free to change this to any name you'd like
data_name = data_path[:-5]
# This is just the default google drive directory, if you wanted to save it to a specific folder in your google drive, it would look like "gdrive/MyDrive/folder/"
google_drive_save_location = "gdrive/MyDrive/"
```

```python
data = pd.read_excel(data_path)
```

```python
data
```

```python
grouped_data = data.groupby(['LFZone', 'EcoSys'])
top_five_frequency = grouped_data.apply(lambda x: x.nlargest(5, frequency_column)).reset_index(drop=True)
top_five_cov = grouped_data.apply(lambda x: x.nlargest(5, abs_cov_column)).reset_index(drop=True)
```

```python
top_five_frequency
```

```python
top_five_cov
```

Exports

```python
top_five_frequency.to_excel(google_drive_save_location + data_name+ "_frequency_result.xlsx")
top_five_cov.to_excel(google_drive_save_location + data_name+ "_cov_result.xlsx")
```
